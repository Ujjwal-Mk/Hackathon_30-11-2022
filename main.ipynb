# QUESTION 3:

Dealing with missing data

```python
#Cleaning Active Column
df["Active"].mean(),df["Active"].median(),df["Active"].mode()
df["Active"]=df["Active"].fillna(df["Active"].mean())
df["Active"].info()
```

```python
#cleaning Rest Column
df["Rest"].isna().sum()
df["Rest"].mean(),df["Rest"].median(),df["Rest"].mode()
df["Rest"]=df["Rest"].fillna(df["Rest"].median())
df["Rest"].info()
```

```python
#Cleaning Smoke column
print(df["Smoke"].isna().sum())
df["Smoke"].mode()
df["Smoke"]=df["Smoke"].replace(np.nan,df["Smoke"].mode()[0])
df["Smoke"].info()
```

```python
#cleaning gender column
print(df["Gender"].isna().sum())
df["Gender"].mode()
df["Gender"]=df["Gender"].replace(np.nan,df["Gender"].mode()[0])
df["Gender"].info()
```

```python
#cleaning Exercise column
print(df["Exercise"].isna().sum())
df["Exercise"].mode()
df["Exercise"]=df["Exercise"].replace(np.nan,df["Exercise"].mode()[0])
df["Exercise"].info()
```

```python
#cleaning height column
df["Hgt"].mean(),df["Hgt"].median(),df["Hgt"].mode()
df["Exercise"]=df["Exercise"].fillna(df["Exercise"].median())
df["Exercise"].info()
```

```python
#cleaning wgt column
df["Wgt"].mean(),df["Wgt"].median(),df["Wgt"].mode()
df["Wgt"]=df["Wgt"].fillna(df["Wgt"].median())
df["Wgt"].info()
```

All the columns have been cleaned

# QUESTION 8:

```python
df_active=df["Active"]
df_rest=df["Rest"]
X_train,X_test,y_train,y_test=train_test_split(df_active,df_rest,test_size=.15,random_state=3)
LR = linear_model.LinearRegression()
train_x=np.asanyarray(X_train)
train_y=np.asanyarray(y_train)
LR.fit(train_x.reshape(-1, 1),train_y)
```

```python
print ('Coefficients: ', LR.coef_)
print ('Intercept: ',LR.intercept_)
test_x=np.asanyarray(X_test)
test_y=np.asanyarray(y_test)
pred_y=LR.predict(test_x.reshape(-1,1))
```

```python
pred_y
```

```python

plt.scatter(test_x,test_y)
plt.plot(train_x, LR.coef_*train_x + LR.intercept_, '-r')
```

```python
mae= np.mean(np.absolute(test_y - pred_y))
mae_=mae/21613
mse=np.mean((test_y - pred_y) ** 2)
mse_=mse/len(test_y)
print("Mean absolute error: " , mae_)
print("Residual sum of squares (MSE): " ,mse_ )
```

From the above error calculations we can see that the **model has high accuracy**

Mean squared error and Mean absolute error(almost 0) is less, hence model fits best

```python
corr_matrix = np.corrcoef(test_y, pred_y)
corr = corr_matrix[0,1]
R_sq = corr**2
 
print("R_squared =",R_sq)
```
